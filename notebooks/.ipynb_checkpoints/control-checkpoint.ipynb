```python
import sys; sys.path.append("../src2")
from pendulum import *
import numpy as np, matplotlib.pyplot as plt
from scipy.optimize import root
```

```python
class controller(object):
    
    def __init__(self, kx=0.25, kv=0.4, kt=13, kw=1.6, um=1):
        self.tr = 0
        self.kx = kx
        self.kv = kv
        self.kt = kt
        self.kw = kw
        self.ul = list()
        self.trl = [self.tr]
        self.um = um
        
    def gen_tt(self, tr):
        xm = np.pi**2*(np.pi + self.um)
        vm = np.pi*self.um
        eq = lambda tt: self.kt*(tt[0]-tr) + np.sin(tt[0]) + self.kt*np.arctan(self.kx*xm + self.kv*vm)
        res = root(eq, [0])
        return res.x[0]
    
    def gen_tr(self, t, tr):
        tt = self.gen_tt(tr)
        dt = t - tr
        if -np.pi <= dt and dt < -tt:
            return tr - 2*np.pi
        elif -tt <= dt and dt <= tt:
            return tr
        elif tt < dt and dt <= np.pi:
            return tr + 2*np.pi
        else:
            return tr
        
    def gen_u(self, x, v, t, w, tr):
        u = (self.kt*(t-tr) + self.kw*w + np.sin(t) + self.kt*np.arctan(self.kx*x + self.kv*v))/np.cos(t)
        u = np.clip(u, -self.um, self.um)
        return u
    
    def __call__(self, state):
        x, v, t, w = state
        u = self.gen_u(x, v, t, w, self.tr)
        return u
    
    def control(self, state):
        x, v, t, w = state
        u = self(state)
        self.tr = self.gen_tr(t, self.tr)
        self.ul.append(u)
        self.trl.append(self.tr)
        return u
        
        
```

```python
cont = controller(kx=0.1, kv=0.1, kt=10, kw=10, um=2)
x0 = [
    0,
    np.random.uniform(-2,2),
    np.random.uniform(0.8*np.pi, 1.2*np.pi),
    np.random.uniform(-1,1)
]
x0 = [0,0,np.pi,0]
dyn  = dynamics(x0,[0,0,0,0], 0)
```

```python
t, x, u = dyn.propagate_controlled(20, cont)
```
