```python
import numpy as np, matplotlib.pyplot as plt
from scipy.integrate import solve_bvp
```

```python
# homotopy parameter
alpha = 0

# fullstate dynamics
def fun(times, fullstates):
    
    # states
    x     = fullstates[0,:]
    v     = fullstates[1,:]
    theta = fullstates[2,:]
    omega = fullstates[3,:]
    
    # costates
    lx     = fullstates[4,:]
    lv     = fullstates[5,:]
    ltheta = fullstates[6,:]
    lomega = fullstates[7,:]
    
    # Pontryagin's minimum principle
    if alpha == 1:
        u = -np.sign(lomega*np.cos(theta)-lv)
    else:
        u = (-lomega*np.cos(theta) + lv)/(2*(alpha - 1))
    
    # common subexpression elimination
    e0 = np.sin(theta)
    e1 = np.cos(theta)
    
    # state dynamics
    dx     = v
    dv     = u
    dtheta = omega
    domega = -u*e1 + e0
    
    # costate dynamics
    dlx     = np.zeros(fullstates.shape[1])
    dlv     = -lx
    dltheta = -lomega*(u*e0 + e1)
    dlomega = -ltheta
    
    # return fullstate transiton
    return np.vstack((dx, dv, dtheta, domega, dlx, dlv, dltheta, dlomega))
```

```python
# fullstate Jacobian
def fun_jac(times, fullstates):
    
    # states
    x     = fullstates[0,:]
    v     = fullstates[1,:]
    theta = fullstates[2,:]
    omega = fullstates[3,:]
    
    # costates
    lx     = fullstates[4,:]
    lv     = fullstates[5,:]
    ltheta = fullstates[6,:]
    lomega = fullstates[7,:]
    
    # Pontryagin's minimum principle
    if alpha == 1:
        u = -np.sign(lomega*np.cos(theta)-lv)
    else:
        u = (-lomega*np.cos(theta) + lv)/(2*(alpha - 1))
        
    # subexpression elimination
    e0 = np.cos(theta)
    e1 = np.sin(theta)
    x2 = u*e1
    
    # blank Jacobian
    n, m = fullstates.shape
    jac = np.zeros((n,n,m))
    
    # set values
    jac[0, 1, :] = np.ones(n)
    jac[2, 3, :] = np.ones(n)
    jac[3, 2, :] = e0 + e2
    jac[5, 4, :] = -np.ones(n)
    jac[6, 2. :] = -lomega*(u*e0 - e1)
    jac[6, 7, :] = -e0 - e2
    jac[7, 6, :] = -np.ones(n)
    
    # return Jacobian
    return jac
```

```python
# boundary constraints
x0 = np.array([0,0,np.pi,0], float)
xf = np.array([0,0,0,0], float)

# boundary residuals
def bc(fullstate0, fullstatef):
    
    # initial state mismatch
    ec0 = fullstate0[:4] - x0
    
    # final state mismatch
    ec1 = fullstatef[:4] - xf
    
    # return constraint vector
    return np.hstack((ec0, ec1))
```

```python
# initial time mesh
nn = 100
timesg = np.linspace(0, 9, nn)
timesg.shape
```

```python
# initial state mesh
fullstatesg = np.vstack((
    *(np.linspace(l, u, nn) for l, u in zip(x0, xf)),
    np.random.uniform(-1,1,(4, nn))
))
fullstatesg.shape
```

```python
res = solve_bvp(fun, bc, timesg, fullstatesg, tol=1e-10, verbose=2, max_nodes=10000)
```
